```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
linear = pd.read_csv('lineara.csv')
linear.head()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.title("Given Dataset",fontsize='16')
plt.grid()
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.show()
```

```python
data1 = pd.read_csv('1guess.csv')
data1.head()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data1.x, data1.y,c='blue')
plt.title("First Guess",fontsize='16')
plt.grid()
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.show()
```

```python
data2 = pd.read_csv('2guess.csv')
data2.head()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data2.x, data2.y,c='green')
plt.title("Second Guess",fontsize='16')
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.grid()
plt.show()
```

```python
data3 = pd.read_csv('3guess.csv')
data3.head()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data3.x, data3.y,c='purple')
plt.title("Third Guess",fontsize='16')
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.grid()
plt.show()
```

```python
data_best= pd.read_csv('best.csv')
data_best.head()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data_best.x, data_best.y,c='black')
plt.title("Graph of Best m and c",fontsize='16')
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data1.x, data1.y,c='blue')
plt.plot(data2.x, data2.y,c='green')
plt.plot(data3.x, data3.y,c='purple')
plt.title("AllThree Guesses",fontsize='16')
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(linear.x, linear.y,c='r')
plt.plot(data1.x, data1.y,c='blue')
plt.plot(data2.x, data2.y,c='green')
plt.plot(data_best.x, data_best.y,c='black')
plt.plot(data3.x, data3.y,c='purple')
plt.xlabel('X-axis',fontsize='15')
plt.ylabel('Y-axis',fontsize='15')
plt.title("Three Guesses with best M and C",fontsize='16')
plt.grid()
plt.show()
```
